```python
import yaml
```

```python
import re
```

```python
EXPORT_UNIT_JSON = re.compile(r"var exportUnitJSON = (.*?)/\*\*", re.MULTILINE|re.DOTALL)
```

```python
with open("./testpage.html") as f:
    data = f.read()
```

```python
match = EXPORT_UNIT_JSON.search(data)
```

```python
data_units = yaml.safe_load(match.group(1))
```

```python
def print_data_unit(collection, level=0):
    for data_unit in collection:
        print('\t' * level, data_unit["text"])
        if "children" in data_unit and data_unit["children"]:
            print_data_unit(data_unit["children"], level=level+1)
```

```python
print_data_unit(data_units)
```
